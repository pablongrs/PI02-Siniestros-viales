```python
def data_null(df):
    list =[]
    for nameCol in df.columns:  
           
        cantidad_nulos = df[nameCol].isnull().sum() #Cuenta los nulos
        porcentaje = round((cantidad_nulos / len(df)) *100) #Calcula el porcentaje de nulos
        tipo_dato = df[nameCol].dtypes #Guarda el tipo de dato
        
    # Por cada iteracion agrega los datos a la lista 
        list.append({'Columna':nameCol,
                     'Tipo Dato':tipo_dato,
                     'Nulos':cantidad_nulos,
                     'Porcentaje':f"{porcentaje} %"
                     })
        
    return pd.DataFrame(list)
        
# cantidad de cadenas vacias por columna
def emptyStrings(df):
    print('FILAS DUPLICADAS: ',df.duplicated().sum())
    print("COLUMNAS CON SD: ")
    for columna in df.columns:
        total = (df[columna] =='SD').sum()
        print(f"{columna} = {total}")  

#Modifica el nombre de cada columna dejando solo la primer letra mayuscula    
def standardized_names(df):
    df.columns = df.columns.str.capitalize()
    return df 

#Se pasa como Argumento una lista de columnas e imprime los datos unicos de C/U
def datosUnicos(df,list):
    for columna in list:
      print(f"Datos unicos {columna}:\n {(df[columna].unique())}")
```

#### Imputacion de datos en columna Longitud y Latitud

Se utilizara libreria **GEOPY** para obtener las coordenadas. Tomando las columnas 'comuna' y 'Calle' se creara una direccion concatenando ambas. La dirección resultante se pasa al geolocalizador de **OpenStreetMap** para obtener las coordenadas de longitud y latitud correspondientes. Tanto la longitud como la latitud no seran las coordenadas exactas donde ocurrio el siniestro. 

```python
#Obtengo solo las filas donde 'Latitud' es NaN
latLong_null = siniestros[siniestros['Latitud'].isna()]

#Se crea un dataframe solamente con las columnas 'comuna' y 'Calle'
latLong = latLong_null[['Comuna','Calle']].copy()
latLong.head(6)
```

```python
#Cambio los NaN por 0
latLong['Comuna'].fillna(0, inplace=True)

#Cambio el tipo de dato a entero
latLong['Comuna'] = latLong['Comuna'].astype('int')
latLong.head()
```

```python

from geopy.geocoders import Nominatim
lista =[] #Se alamcenara los datos obtenidos de la api (longitud y latitud)

geolocator = Nominatim(user_agent="my_geocoder")

#Itera sobre cada fila del DF 'latLong'
for fila, valorFila in latLong.iterrows():
    
    # Verifica si el valor de la fila actal en 'Comuna' es mayor a 0
    if (valorFila["Comuna"] > 0):
        # Verifica que el valor de la fila actual en 'Calle' sea texto
        if isinstance(valorFila["Calle"], str):            
          try:
              #se le asigna a la variable un txt con la comuna y calle. Ej: Comuna 9, MURGUIONDO
              direccion = 'Comuna {}, {}'.format(valorFila["Comuna"], valorFila["Calle"])
              location = geolocator.geocode(direccion) #Toma como argumento la direccion y obtiene la ubicacion
              #if location:
              lista.append({ 'Fila':fila,
                               'Longitud': location.longitude,
                                'Latitud': location.latitude})                               
          except:
              print(f"Indice {fila}. ocurrio un error al obtener los datos")
         
```

```python
df_longitud_latitud = pd.DataFrame(lista)
df_longitud_latitud = df_longitud_latitud.set_index('Fila')

df_longitud_latitud.head()
```

```python
indices_lista = df_longitud_latitud.index.tolist()

#Selecciona las filas que coinciden con los valores de 'indices_lista' y en las columnas 'Longitud Latitud'
#actualiza  con los valores correspondientes del DataFrame 'df_longitud_latitud'
siniestros.loc[indices_lista, ['Longitud', 'Latitud']] = df_longitud_latitud[['Longitud', 'Latitud']].values
```
